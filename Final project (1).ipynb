### Data Cleaning

#### 1. Droping the Data`

```python
p_rta.drop(["Time","Service_year_of_vehicle","Work_of_casuality","Fitness_of_casuality",
              "Day_of_week","Educational_level","Defect_of_vehicle","Sex_of_driver",
              "Casualty_class","Sex_of_casualty","Age_band_of_casualty","Casualty_severity","Pedestrian_movement"],
             axis=1,inplace=True)
```

**Insights**
- The columns Service_year_of_vehicle, Defect_of_vehicle, Work_of_casuality, and Fitness_of_casuality were dropped due to high missing values and low data reliability.
- Columns such as Time and Day_of_week were removed as they are not relevant to the defined analysis scope.
- Variables like Educational_level and Sex_of_driver were excluded due to low impact on accident severity analysis.
- Detailed casualty-related fields including Casualty_class, Sex_of_casualty, and Age_band_of_casualty were dropped to avoid unnecessary complexity.
- Finally, Casualty_severity and Pedestrian_movement were removed due to redundancy and lack of relevance to the project objectives.

#### 2. Handling Missing Data

```python
df=p_rta.isna().sum().sort_values(ascending=False)/len(p_rta)*100
df
```

#### Fill missing values in each column

```python
# Forward Fill (ffill)
p_rta["Type_of_vehicle"] = p_rta["Type_of_vehicle"].ffill()
p_rta["Area_accident_occured"] = p_rta["Area_accident_occured"].ffill()
```

```python
#Fill with "Unknown"
p_rta["Types_of_Junction"] = p_rta["Types_of_Junction"].fillna("Unknown")
p_rta["Driving_experience"] = p_rta["Driving_experience"].fillna("Unknown")
p_rta["Lanes_or_Medians"] = p_rta["Lanes_or_Medians"].fillna("Unknown")
```

```python
# Fill with "No Value"
p_rta["Vehicle_movement"] = p_rta["Vehicle_movement"].fillna("No Value")
p_rta["Type_of_collision"] = p_rta["Type_of_collision"].fillna("No Value")
p_rta["Road_allignment"] = p_rta["Road_allignment"].fillna("No Value")
```

```python
# Backward Fill (bfill)
p_rta["Vehicle_driver_relation"] = p_rta["Vehicle_driver_relation"].bfill()
p_rta["Road_surface_type"] = p_rta["Road_surface_type"].bfill()
```

```python
# Fill mode (most frequent value) for categorical column
p_rta["Owner_of_vehicle"] = p_rta["Owner_of_vehicle"].fillna(
    p_rta["Owner_of_vehicle"].mode()[0]
)
```

```python
df=p_rta.isna().sum().sort_values(ascending=False)/len(p_rta)*100
df
```

**Insights**:
- Different missing value techniques were applied based on the nature and importance of each column to preserve data quality.
- Forward fill (ffill) was used for columns like Type_of_vehicle and Area_accident_occured to maintain continuity in similar accident records.
- Columns with unclear or unavailable information were filled using labels like “Unknown” or “No Value” to avoid losing valuable rows.
- Backward fill (bfill) and mode imputation were used to replace missing categorical values with the nearest or most frequent meaningful category.
- This preprocessing step reduced missing data effectively, improved dataset consistency, and prepared the data for accurate analysis and machine learning.

## Data Visualization & EDA

### 1. Impact of Weather on Accident Frequency Using Count Plot

```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=p_rta,
    x='Weather_conditions',
    order=p_rta['Weather_conditions'].value_counts().index,
    palette='magma'
)

plt.title('Accidents by Weather Condition', fontsize=16, fontweight='bold')
plt.xlabel('Weather Conditions', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)

plt.xticks(rotation=45)
plt.show()
```

**Insight**:
- Most accidents occurred during normal weather conditions compared to other weather types.
- However, the data also shows that a significant number of accidents happened during rainy weather, indicating that weather conditions can still impact road safety and driving conditions.


### 2. Severity Level Distribution of Accidents Using Pie Chart Plot

```python
Accident_severity = p_rta['Accident_severity'].value_counts() # Count Accident Severity
colors = [ 'orange','teal', 'blue'] # Colors for Pie Chart

plt.figure(figsize=(8,8))

plt.pie(
    Accident_severity,
    labels=Accident_severity.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    shadow=False,
    explode=(0.05, 0.05, 0.05)
)

plt.title('Percentage Distribution of Accident Severity',
    fontsize=16,
    fontweight='bold'
)
plt.show()
```

**Insight**:
- As we can see, the majority of accidents resulted in slight injuries, accounting for 84.6% of the total cases. 
- Serious injuries made up 14.2% of the accidents, while fatal injuries accounted for only 1.3%, indicating that severe and fatal accidents were comparatively lower.

### 3. Area Wise Accident Trends Using Line Plot

```python
area_accidents = p_rta["Area_accident_occured"].value_counts()
plt.figure(figsize=(12,6))

plt.plot(area_accidents.index,
    area_accidents.values,
    'r-.d',   # yellow color, dash-dot line, diamond marker
    linewidth=3
)
plt.title(
    'Area-wise Number of Accidents',
    fontsize=16,
    fontweight='bold'
)
plt.xlabel('Area Where Accident Occurred', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=90)
plt.grid(False)
plt.show()
```

**Insight**:
- The highest number of accidents occurred in “Other” and office areas, followed by residential areas. 
- In contrast, rural villages, market areas, and recreational areas recorded comparatively fewer accidents.

### 4. Most Common Collision Types Using Bar Chart Plot

```python
collision_counts = p_rta['Type_of_collision'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=collision_counts.index, y=collision_counts.values, palette="viridis")
plt.title("Most Frequent Collision Types")
plt.xlabel("Type of Collision")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=90)
plt.show()

```

**Insight**: 
- The chart shows that vehicle-to-vehicle collisions dominate accident frequency, making them the primary focus area for road safety interventions.
- In contrast, collisions with pedestrians and roadside objects are the next most common, highlighting critical risks for vulnerable road users and infrastructure.

### 5. Urban vs Rural Accident Split Using Historam

```python
plt.figure(figsize=(10,6))
sns.histplot(p_rta["Area_accident_occured"],bins=10,kde=False,color="Purple")
plt.title("Accident split between Urban and Rural Areas")
plt.xlabel("Urban Areas")
plt.ylabel("Rural Areas")
plt.xticks(rotation=90)
plt.show()
```

**Insight**: 
- Accidents are most concentrated in office, residential, and “other” urban areas, showing that dense activity zones face higher risks.
- In contrast, rural and specialized locations like markets or churches record far fewer incidents, indicating lower exposure but still notable vulnerability.

### 6. Driver Age Bands and Accident Distribution Using Pie Chart Plot

```python
area_counts = p_rta['Age_band_of_driver'].value_counts()
plt.figure(figsize=(10,6))
plt.pie(area_counts, labels = area_counts.index, autopct='%1.1f%%', 
        startangle=90, colors=['Red','lightgreen','lightblue','orange','pink'])
plt.title("Distribution of accident counts across different age bands of drivers")
plt.show()
```

### 7. Vehicle Types Most Involved in Accidents Using Horizontal Bar Graph Plot

### 8. Accident Density Across Districts/Areas Using Heat Map Plot

```python
accident_density = pd.crosstab(
    p_rta['Area_accident_occured'],
    p_rta['Accident_severity']
)

plt.figure(figsize=(12,6)) # Plot heatmap

sns.heatmap(
    accident_density,
    annot=True,
    cmap='Reds',
    fmt='d'
)
plt.title("Accident Density by Area and Severity")
plt.xlabel("Accident Severity")
plt.ylabel("Area of Accident")

plt.show()
```

**Insights**:
- Office areas recorded the highest number of accidents
- Most accidents fall under lower to medium severity level
- Industrial and school areas have comparatively fewer accidents

### 9. Patterns Across Accident Variables Using Scatter Plot

**Insights**:

### 12. Weather Conditions vs Road Surface Conditions Using Heatmap Plot

```python
plt.figure(figsize=(10,6))

cross_tab = pd.crosstab(
    p_rta['Weather_conditions'],
    p_rta['Road_surface_conditions']
)

sns.heatmap(
    cross_tab,
    annot=True,
    cmap="YlOrRd", # Color Yl= Yellow, Or=Orange, Rd= Red
    fmt='d' # Display values for interger
)

plt.title("Weather Conditions vs Road Surface Conditions")

plt.show()
```

**Insights**:
- Most accidents occurred during normal weather on dry roads, indicating that accident frequency is influenced more by traffic volume than extreme weather conditions.
- Rainy weather is strongly associated with wet or damp road surfaces, suggesting that slippery roads significantly contribute to accident occurrences.
